```python
import psutil
import humanize
import os
import GPUtil as GPU
```

```python
GPUs = GPU.getGPUs()
gpu = GPUs[0]
```

```python
print(len(GPUs))
```

```python
print(
    '''GPU RAM Free: {}MB | Used: {}MB | Util {}% | Total {}MB'''.format(
        gpu.memoryFree, gpu.memoryUsed, gpu.memoryUtil, gpu.memoryTotal
    )
)
```

```python
# find the RAM available to the system use : psutil.virtual_memory().available

print("Gen RAM Free: " + humanize.naturalsize(psutil.virtual_memory().available))
```

```python
# find how much process is used : process.memory_info() tells us how much memory is concuming
process = psutil.Process(os.getpid())
print("Gen Used : ", humanize.naturalsize(process.memory_info().rss))
```
